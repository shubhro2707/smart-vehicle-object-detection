```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

#load the image from the data folder
img = cv2.imread('../data/road.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #for correct color in matplotlib

#grayscale conversion
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#math Y = 0.299R + 0.587G + 0.114B

#gaussian blur
blurred = cv2.GaussianBlur(gray, (5,5),0)
#each pixel = weighted average of its neighbors via gaussian kernel

#edge dection using canny
edges = cv2.Canny(blurred, 50, 150)
#uses gradient, sobel and hysterisis thresholding

#plot all images side by side
fig, axs = plt.subplots(1, 4, figsize = (20,5))
axs[0].imshow(img_rgb)
axs[0].set_title('original')

axs[1].imshow(gray)
axs[1].set_title('Grayscale')

axs[2].imshow(blurred)
axs[2].set_title('Gaussian Blur')

axs[3].imshow(edges)
axs[3].set_title('Canny Edges')

for ax in axs:
    ax.axis('off')
plt.tight_layout()
plt.show()
```
